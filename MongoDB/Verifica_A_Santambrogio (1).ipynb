# i consolati che non hanno una mail del Ministero degli Esteri

```python
result = db.consolati.find({"email":{"$not":{"$regex": "@esteri.it"}}})
df9 = pd.DataFrame(list(result))
df9
```

# il numero di consolati in ogni continente

```python
result = db.consolati.aggregate([
                                 {"$group": {"_id" : "$continente", "total" : {"$sum" : 1}}},
                                 {"$sort": {"_id": 1}}                         
])
df10 = pd.DataFrame(list(result))
df10
```

# i continenti che hanno più di 50 consolati. Ordinare in ordine decrescente a partire dal continente che ha più consolati

```python
result = db.consolati.aggregate([
                                 {"$group": {"_id" : "$continente", "total" : {"$sum" : 1}}},
                                 {"$match":{"total":{"$gt": 50}}},
                                 {"$sort":{"total": -1}}
])
df11 = pd.DataFrame(list(result))
df11
```

# il paese che ha più consolati

```python
result = db.consolati.aggregate([
                                 {"$group": {"_id" : "$paese", "total" : {"$sum" : 1}}},
                                 {"$sort":{"total": -1}}
])
df12 = pd.DataFrame(list(result))
df12
```
