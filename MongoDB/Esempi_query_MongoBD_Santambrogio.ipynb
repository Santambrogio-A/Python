#Funzioni di aggregazione, group by e having

```
# conta quanti documenti ci sono nella collection restaurant
result = db.restaurants.aggregate([
                                   {"$group" : {"_id": "null", "count" : {"$sum" : 1}}}
                                   ])
df2 = pd.DataFrame(list(result))
df2.head()
```

```
# conta i ristoranti di ogni quartiere
result = db.restaurants.aggregate([
                                   {"$group" : {"_id": "$borough", "total" : {"$sum" : 1}}}
                                   ])
df2 = pd.DataFrame(list(result))
df2.head()
```

```
# conta i ristoranti di ogni quartiere e li inserisce in ordine decrescente
result = db.restaurants.aggregate([ 
                                   {"$group" : {"_id": "$borough", "total" : {"$sum" : 1}}} , 
                                   { "$sort": { "total": -1 } } 
                                   ])
df2 = pd.DataFrame(list(result))
df2.head()
```

```
# prendiamo i quartieri con un numero di ristoranti maggiori di 3000 e li oridiniamo in ordine necrescente 
result = db.restaurants.aggregate([ 
                                   {"$group" : {"_id": "$borough", "total" : {"$sum" : 1}}} , 
                                   { "$match": { "total": { "$gt": 3000 } } }, 
                                   { "$sort": { "total": -1 } } 
                                   ])
df2 = pd.DataFrame(list(result))
df2.head()
```

```
# prendiamo i quartieri con un numero di ristoranti con cucina italiana maggiori di 70 e li oridiniamo in ordine decrescente 
result = db.restaurants.aggregate([ 
                                   { "$match": { "cuisine": "Italian" } }, 
                                   { "$group" : {"_id": "$borough", "total" : {"$sum" : 1}}} , 
                                   { "$match": { "total": { "$gt": 70 } } }, 
                                   { "$sort": { "total": -1 } } ])
df2 = pd.DataFrame(list(result))
df2.head()
```
