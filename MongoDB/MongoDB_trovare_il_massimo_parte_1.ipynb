```python
# Fare un $group per trovare il massimo sul campo customer.age, utilizzando il push inseriamo i dati all'interno di un nuovo campo che si chiama customerdata
result = db2.venditemax.aggregate([{ "$group": {"_id": "maxAge","maxAge": {"$max": "$customer.age"},
        "customerdata": {"$push": {"age": "$customer.age", "email": "$customer.email"}}}}])
df = pd.DataFrame(list(result))
df
```

```python
#Fare un unwind per estrarre i singoli customer dalla lista customerdata
result = db2.venditemax.aggregate([{ "$group": {"_id": "maxAge","maxAge": {"$max": "$customer.age"},
        "customerdata": {"$push": {"age": "$customer.age", "email": "$customer.email"}}}},{'$unwind': '$customerdata'}])
df = pd.DataFrame(list(result))
df
```

```python
# Fare un confornto tra maxAge e customerdata.age per vedere se sono uguali 
# utilizziamo l'operatore $match 
result = db2.venditemax.aggregate([
     
    { "$group": {
        "_id": "maxAge",
        "maxAge": {
            "$max": "$customer.age"
        },
        "customerdata": {
            "$push": {
                "age": "$customer.age",
                "email": "$customer.email"
            }
        }}
       },
       
      {'$unwind': '$customerdata'
    },
    {'$match':{'$expr':{"$eq":["$maxAge", "$customerdata.age"]}}}
    ])
df = pd.DataFrame(list(result))
df
```
