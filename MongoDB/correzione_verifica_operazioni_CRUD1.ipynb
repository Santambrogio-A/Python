#7. visualizare il nome del ristorante con il restaurant_id piu basso

```python
result = db2.ristoranti.aggregate([
    { "$group": {
        "_id": "",
        "ResIdMin": {
            "$min": "$restaurant_id"
        },
        "restaurantIDPiuBasso": {
            "$push": {
                "nome": "$name",
                "id": "$restaurant_id"
            }
        }
       }
    },
    {"$unwind" : "$restaurantIDPiuBasso"},
    {"$match": {"$expr": {"$eq": [ "$restaurantIDPiuBasso.id","$ResIdMin"]}}}])

df = pd.DataFrame(list(result))
df
```

#8. trovare i tipi di cucina piu diffusi a Brooklyn
